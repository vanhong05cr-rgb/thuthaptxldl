## 1.IMPORT THƯ VIỆN

```python
import pandas as pd
df = pd.read_csv("IMDB-Movie-Data.csv")
```

## 2. VIEW THE DATA


```python
df.head()       # Xem 5 dòng đầu tiên
df.shape        # Số dòng và số cột
df.info()       # Thông tin kiểu dữ liệu
df.describe()   # Thống kê cơ bản
```

```python
df[['Title', 'Rating']].head(10)  # Lấy 10 phim đầu tiên gồm Tên và Điểm
df.iloc[0:5, 0:3]                 # Lấy 5 dòng đầu và 3 cột đầu
```

```python
df[df['Rating'] >= 8.0][['Title', 'Rating']].head(10)  # Phim rating ≥ 8
```

```python
df.groupby('Year')['Rating'].mean().head()   # Trung bình rating theo năm
df.groupby('Director')['Revenue (Millions)'].mean().sort_values(ascending=False).head(5)  
```

```python
df.sort_values(by='Rating', ascending=False).head(10)   # Top 10 phim theo rating
df.sort_values(by='Revenue (Millions)', ascending=False).head(10) # Top 10 phim doanh thu
```

```python
df.isnull().sum()     # Đếm số lượng giá trị thiếu
df_drop = df.dropna() # Xóa dòng thiếu dữ liệu
df_fill = df.fillna(df.mean(numeric_only=True))  # Điền bằng trung bình
```

```python
df['Title_Upper'] = df['Title'].apply(lambda x: x.upper())
df[['Title', 'Title_Upper']].head()
```

```python
import matplotlib.pyplot as plt

# Histogram Rating
plt.hist(df['Rating'], bins=20, color='skyblue', edgecolor='black')
plt.title("Phân bố điểm IMDB")
plt.xlabel("Rating")
plt.ylabel("Số lượng phim")
plt.show()

```
